````python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import imutils
from imutils.object_detection import non_max_suppression
````

````python
img = plt.imread("instreet.jpeg")
````

````python
plt.imshow(img)
````

### Firstly Create HOG descriptor for each person

````python
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
````

### Rectangle of detected objects

````python
(r,w) = hog.detectMultiScale(img, winStride=(2,2), padding=(2,2), scale=1.1)
````

````python
rects = np.array([[x, y, x+w, y+h] for (x, y, w, h) in r])
````

````python
rects_NMS = non_max_suppression(rects, probs=None, overlapThresh=0.8)
````

````python

for (x1, y1, x2, y2) in rects_NMS:
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 1)
````
